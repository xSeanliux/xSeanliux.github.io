```python
from unidecode import unidecode
from os import listdir
from os.path import isfile, join
import re
from PIL import Image
```

```python
dir = "./cuneiform_images/"
filenames = [f for f in listdir(dir) if isfile(join(dir, f))]
```

```python
all_names = [(x, unidecode(x).lower()) for x in filenames]
```

```python
with open("./cuneiform_names.txt", "w") as f:
    for i, (filename, key) in enumerate(all_names):
        res = re.match("[a-z]*", key)
        print(filename, key, res[0])
        f.write(f"{res[0]}\t{filename}\n")
    f.write("hursag HUR.SAG.png")
    f.write("hur.sag HUR.SAG.png")
```

```python
import numpy as np
heights = []

for filename in filenames:
    with Image.open(dir + filename) as im:
        heights.append(im.height)

print("average height: ", np.array(heights).mean())

target_height = int(round(np.array(heights).mean()))

print("target height: ", target_height)

for filename in filenames:
    with Image.open(dir + filename) as im:
        im_resized = im.resize((round(im.width * target_height / im.height), round(target_height)))
        print(im_resized.width, im_resized.height)
        im_resized.save("./cuneiform_images_resized/" + filename)
```
